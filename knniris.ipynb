```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
import matplotlib as plt
```

```python
df=pd.read_csv("/content/Iris (1).csv")
df.describe()
```

```python
X=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=df['Species']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
print("X_train{}\ny_train{}".format(X_train.shape, y_train.shape))
print("X_test{}\ny_test{}".format(X_test.shape, y_test.shape))
```

```python
model= KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X_train, y_train)
predictions=model.predict(X_test)
print(model.score(X_test,y_test))
```

```python
pd.concat([X_test, y_test, pd.Series(predictions, name='valeurs predites', index=X_test.index)],ignore_index=False, axis=1)
```
